```python
import tensorflow
import numpy as np
from tensorflow import keras

model = keras.Sequential(keras.layers.Dense(units=1, input_shape=[1]))

model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([float(x) for x in np.arange(1, 8)], dtype=float)
ys = np.array([(float(x) for x in np.arange(1, 8, 0.5))], dtype=float)

model.fit(xs, ys, epochs=500)
```

```python
model.predict([7.0])
```
